```python
csvfile = open('Admission.csv', 'r')
```

```python
reader = csv.DictReader( csvfile )
```

```python
header= [ 'Serial No.','GRE Score','TOEFL Score','University Rating','SOP','LOR ' ,'CGPA','Research','Chance of Admit ' ]

for each in reader:
    row={}
    for field in header:
        row[field]=each[field]

    collection.insert_one(row)
```

```python
res=collection.find()
```
